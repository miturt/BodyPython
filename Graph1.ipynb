### 時間軸の拡大

```python
##############################################
start = 12.5
end = 17.5
ku = 1
###############################################

nst,ned = startend(da, start, end , ku)
# print("nst", nst)
# print("ned", ned)

for k in range(ku):
#     print('k ',k)

    fig = plt.figure(figsize=(10, 12))

    ax1 = fig.add_subplot(211)
    ax1.set_title(head + "   " + commt)

    ax1.set_xlim(time[nst[k]],time[ned[k]])
    ax1.plot(da[0,], da[1,], color="red", label=names[1])
    ax1.plot(da[0,], da[2,], color="green", label=names[2])
    ax1.plot(da[0,], da[3,], color="blue", label=names[3])
    ax1.set_ylabel("Gyroscope")
    ax1.legend()
    plt.grid()

    ax2 = fig.add_subplot(212)
    ax2.set_xlim(time[nst[k]],time[ned[k]])
    ax2.plot(da[0,], da[4,], color="red", label=names[4])
    ax2.plot(da[0,], da[5,], color="green", label=names[5])
    ax2.plot(da[0,], da[6,], color="blue", label=names[6])
    ax2.set_ylabel("Accelation")
    ax2.legend()
    plt.grid()

fpng = fname + '_k1' + '.png'

plt.savefig(fpng,bbox_inches="tight")
plt.show()
```
