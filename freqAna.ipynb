## 周波数分析
高速フーリエ変換（FFT)による周波数分析です。
窓関数としてハニング窓を適用したれいです。
実行する前に　８，９行目のデータファイルの指定、および４１，４２行目の区間の指定を変更してください。

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import scipy.fftpack
from bodycsv import *
%matplotlib inline

path = '/Users/miy/Downloads/'
fname = '20190604_1200'
# path = "/Users/miya/data/mdata/20171109_1113.mdata"
fcsv = path + fname + '.csv'

head, commt, names, da = readcsv(fcsv)


aaa = names[0]
ns = locals()
ns[aaa] = da[0,]

# print(head,commt)
# print(names)
# print(len(da[0,]))
# print("time ",da[0,0],max(da[0,]))

# for i in range(4,7):
#     print(names[i], min(da[i,]), max(da[i,]))

time = np.array(da[0,]-da[0,0])
dmax= max(da[6,])
dt = da[0,1]-da[0,0]
# print("dt= ","{:.3F}".format(dt))


# acplane = np.array(da[5,]**2 + da[6,]**2)

# plt.plot(time,acplane)
x = [y/dmax for y in da[6,]]
plt.plot(time,da[6,])
plt.ylabel(names[6])

start = 400 # サンプリングする開始位置
N = 128 * 3 # FFTのサンプル数
fs = 1./dt # サンプリング周波数 100Hz => 10m秒間隔でサンプリング
stime = start/100.
print("start= " + "{:.1f}".format(stime) + " sec")
print("end = " + "{:.1f}".format((start+N)/fs) + " sec")

hanningWindow = np.hanning(N) # ハニング窓
X = scipy.fftpack.fft(hanningWindow * x[start:start+N])
# print('step1')
freqList = scipy.fftpack.fftfreq(N, d=1.0/ fs) # 周波数軸の値を計算
# print('step2')
amplitudeSpectrum = [np.sqrt(c.real ** 2 + c.imag ** 2) for c in X] # 振幅スペクトル
# print('step3')

fig = plt.figure(figsize=(8, 6))

ax1 = fig.add_subplot(211)

# 元の波形
ax1.plot(range(start, start+N), x[start:start+N])
ax1.axis([start, start+N, -1.0, 1.0])
ax1.set_title(head + "  " + commt + "   Original Signal")
ax1.set_xlabel("time [sample]")
ax1.set_ylabel(names[6])

plt.grid()


# 振幅スペクトル
ax2 = fig.add_subplot(212)
ax2.plot(freqList, amplitudeSpectrum, marker= 'o', linestyle='-')
# ax2.axis([0, fs/2, 0, 2])
ax2.set_xlim(0, 12)
ax2.set_xlabel("frequency [Hz]")
ax2.set_ylabel("amplitude spectrum")
plt.grid()

plt.show()


```
